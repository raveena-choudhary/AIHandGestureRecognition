```python
import cv2
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
# import os
import numpy as np
import random
import math

mp_hands = mp.solutions.hands

# Function to draw balloons animation
def draw_balloons(frame, positions):
    colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]  # Example colors
    for i, position in enumerate(positions):
        color = colors[i % len(colors)]
        cv2.circle(frame, position, 20, color, -1)

def victory_detection(frame,hand_landmarks):

    # Check if palm is open
    thumb_cmc = hand_landmarks.landmark[mp_hands.HandLandmark.THUMB_CMC]
    thumb_tip = hand_landmarks.landmark[mp_hands.HandLandmark.THUMB_TIP]
    # index_finger_mcp = hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_MCP]
    # index_finger_tip = hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP]
    # middle_finger_mcp = hand_landmarks.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_MCP]
    # middle_finger_tip = hand_landmarks.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_TIP]
    ring_finger_mcp = hand_landmarks.landmark[mp_hands.HandLandmark.RING_FINGER_MCP]
    ring_finger_tip = hand_landmarks.landmark[mp_hands.HandLandmark.RING_FINGER_TIP]
    pinky_mcp = hand_landmarks.landmark[mp_hands.HandLandmark.PINKY_MCP]
    pinky_tip = hand_landmarks.landmark[mp_hands.HandLandmark.PINKY_TIP]

    # Check if any of the finger landmarks are not detected
    # if any(landmark is not None for landmark in [thumb_cmc, thumb_tip,ring_finger_mcp,
    #                                          ring_finger_tip, pinky_mcp, pinky_tip]):
    #     return frame

    # print(thumb_tip.x)
    # print(ring_finger_tip.x)
    # print(hand_landmarks.landmark[17].x)

    #handling both hands
    if thumb_tip.x >= ring_finger_tip.x:
        if (hand_landmarks.landmark[2].x >= hand_landmarks.landmark[13].x) and (hand_landmarks.landmark[2].x >= hand_landmarks.landmark[17].x):
            return frame
    elif thumb_tip.x < ring_finger_tip.x:
        if (hand_landmarks.landmark[2].x < hand_landmarks.landmark[13].x) and (hand_landmarks.landmark[2].x < hand_landmarks.landmark[17].x):
            return frame
        
    # Extract coordinates of index finger MCP, index finger tip, middle finger MCP, and middle finger tip
    index_finger_mcp = hand_landmarks.landmark[5]
    index_finger_tip = hand_landmarks.landmark[8]
    middle_finger_mcp = hand_landmarks.landmark[9]
    middle_finger_tip = hand_landmarks.landmark[12]

    # print(index_finger_tip, index_finger_mcp)

    # Convert normalized coordinates to pixel coordinates
    h, w, _ = frame.shape
    cx1, cy1 = int(index_finger_mcp.x * w), int(index_finger_mcp.y * h)
    cx2, cy2 = int(index_finger_tip.x * w), int(index_finger_tip.y * h)
    cx3, cy3 = int(middle_finger_mcp.x * w), int(middle_finger_mcp.y * h)
    cx4, cy4 = int(middle_finger_tip.x * w), int(middle_finger_tip.y * h)

    distance_threshold = 40  #pixels
    distance = math.sqrt((cx4 - cx2)**2 + (cy4 - cy2)**2)
            
    # Find the bounding box coordinates
    # min_x = min(cx1, cx2, cx3, cx4)
    # min_y = min(cy1, cy2, cy3, cy4)
    # max_x = max(cx1, cx2, cx3, cx4)
    # max_y = max(cy1, cy2, cy3, cy4)

    # Draw rectangle around the bounding box
     # Check if all required landmarks are found
    if index_finger_mcp is not None and index_finger_tip is not None and \
       middle_finger_mcp is not None and middle_finger_tip is not None:

    #     # Check if the index and middle fingers are up
        if (index_finger_tip.y < index_finger_mcp.y and middle_finger_tip.y < middle_finger_mcp.y) \
           or (index_finger_tip.y > index_finger_mcp.y and middle_finger_tip.y > middle_finger_mcp.y)\
           and distance >= distance_threshold:
            # Draw rectangle around the landmarks
            min_x = min(cx1, cx2, cx3, cx4)
            min_y = min(cy1, cy2, cy3, cy4)
            max_x = max(cx1, cx2, cx3, cx4)
            max_y = max(cy1, cy2, cy3, cy4)
            
            cv2.rectangle(frame, (min_x, min_y), (max_x, max_y), (255, 0, 0), 2)
            
            # Add text to the rectangle
            text = "Victory!"
            font = cv2.FONT_HERSHEY_COMPLEX_SMALL
            font_scale = 1
            font_thickness = 1
            text_size = cv2.getTextSize(text, font, font_scale, font_thickness)[0]
            text_x = min_x + (max_x - min_x) // 2 - text_size[0] // 2
            text_y = min_y + (max_y - min_y) // 2 + text_size[1] // 2
            cv2.putText(frame, text, (text_x, text_y), font, font_scale, (255, 255, 255), font_thickness)

            # positions = [(random.randint(0, frame.shape[1]), random.randint(0, frame.shape[0])) for _ in range(3)]
            # # Draw balloons animation
            # draw_balloons(frame,positions)
            
    return frame

def thumbup_detection(frame,hand_landmarks): 
    thumb_cmc = hand_landmarks.landmark[mp_hands.HandLandmark.THUMB_CMC]
    thumb_tip = hand_landmarks.landmark[mp_hands.HandLandmark.THUMB_TIP]
    index_finger_mcp = hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_MCP]
    index_finger_tip = hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP]
    middle_finger_mcp = hand_landmarks.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_MCP]
    middle_finger_tip = hand_landmarks.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_TIP]
    ring_finger_mcp = hand_landmarks.landmark[mp_hands.HandLandmark.RING_FINGER_MCP]
    ring_finger_tip = hand_landmarks.landmark[mp_hands.HandLandmark.RING_FINGER_TIP]
    pinky_mcp = hand_landmarks.landmark[mp_hands.HandLandmark.PINKY_MCP]
    pinky_tip = hand_landmarks.landmark[mp_hands.HandLandmark.PINKY_TIP]

    # print(thumb_tip.y)
    # print(index_finger_tip.y)

     # Convert normalized coordinates to pixel coordinates
    h, w, _ = frame.shape
    cx1, cy1 = int(thumb_tip.x * w), int(thumb_tip.y * h)
    cx2, cy2 = int(thumb_cmc.x * w), int(thumb_cmc.y * h)

    # Draw rectangle around the bounding box

    #     # Check if thumb is up
    if (thumb_tip.y<index_finger_tip.y and index_finger_tip.y<middle_finger_tip.y and middle_finger_tip.y<ring_finger_tip.y and ring_finger_tip.y<pinky_tip.y):
        # or (thumb_tip.y<index_finger_tip.y and index_finger_tip.y<middle_finger_tip.y and middle_finger_tip.y<ring_finger_tip.y and ring_finger_tip.y<pinky_tip.y):
            # Draw rectangle around the landmarks
            min_x = min(cx1, cx2)
            min_y = min(cy1, cy2)
            max_x = max(cx1, cx2)
            max_y = max(cy1, cy2)
            
            cv2.rectangle(frame, (min_x, min_y), (max_x, max_y), (255, 0, 0), 2)
            
            # Add text to the rectangle
            text = "ok!"
            font = cv2.FONT_HERSHEY_COMPLEX_SMALL
            font_scale = 1
            font_thickness = 1
            text_size = cv2.getTextSize(text, font, font_scale, font_thickness)[0]
            text_x = min_x + (max_x - min_x) // 2 - text_size[0] // 2
            text_y = min_y + (max_y - min_y) // 2 + text_size[1] // 2
            cv2.putText(frame, text, (text_x, text_y), font, font_scale, (255, 255, 255), font_thickness)
            
    return frame

def thumbdown_detection(frame,hand_landmarks): 
    thumb_cmc = hand_landmarks.landmark[mp_hands.HandLandmark.THUMB_CMC]
    thumb_tip = hand_landmarks.landmark[mp_hands.HandLandmark.THUMB_TIP]
    index_finger_mcp = hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_MCP]
    index_finger_tip = hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP]
    middle_finger_mcp = hand_landmarks.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_MCP]
    middle_finger_tip = hand_landmarks.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_TIP]
    ring_finger_mcp = hand_landmarks.landmark[mp_hands.HandLandmark.RING_FINGER_MCP]
    ring_finger_tip = hand_landmarks.landmark[mp_hands.HandLandmark.RING_FINGER_TIP]
    pinky_mcp = hand_landmarks.landmark[mp_hands.HandLandmark.PINKY_MCP]
    pinky_tip = hand_landmarks.landmark[mp_hands.HandLandmark.PINKY_TIP]

    # print(thumb_tip.y)
    # print(index_finger_tip.y)

     # Convert normalized coordinates to pixel coordinates
    h, w, _ = frame.shape
    cx1, cy1 = int(thumb_tip.x * w), int(thumb_tip.y * h)
    cx2, cy2 = int(thumb_cmc.x * w), int(thumb_cmc.y * h)

    # Draw rectangle around the bounding box

    #     # Check if thumb is down
    if (thumb_tip.y>index_finger_tip.y and index_finger_tip.y>middle_finger_tip.y and middle_finger_tip.y>ring_finger_tip.y and ring_finger_tip.y>pinky_tip.y):
        # or (thumb_tip.y<index_finger_tip.y and index_finger_tip.y<middle_finger_tip.y and middle_finger_tip.y<ring_finger_tip.y and ring_finger_tip.y<pinky_tip.y):
            # Draw rectangle around the landmarks
            min_x = min(cx1, cx2)
            min_y = min(cy1, cy2)
            max_x = max(cx1, cx2)
            max_y = max(cy1, cy2)
            
            cv2.rectangle(frame, (min_x, min_y), (max_x, max_y), (255, 0, 0), 2)
            
            # Add text to the rectangle
            text = "not ok!"
            font = cv2.FONT_HERSHEY_COMPLEX_SMALL
            font_scale = 1
            font_thickness = 1
            text_size = cv2.getTextSize(text, font, font_scale, font_thickness)[0]
            text_x = min_x + (max_x - min_x) // 2 - text_size[0] // 2
            text_y = min_y + (max_y - min_y) // 2 + text_size[1] // 2
            cv2.putText(frame, text, (text_x, text_y), font, font_scale, (255, 255, 255), font_thickness)
            
    return frame

#create mp object
hands = mp.solutions.hands.Hands(static_image_mode=False, max_num_hands=2, min_detection_confidence=0.65, min_tracking_confidence=0.65)

# Open the camera
cap = cv2.VideoCapture(1)

while True:
    # Capture frame-by-frame
    valid_frame, frame = cap.read()
    if not valid_frame:
        break

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Process the frame using MediaPipe Hands
    results = hands.process(frame_rgb)

     # Display the results
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            # print(hand_landmarks)
            # Draw landmarks on the frame
            mp.solutions.drawing_utils.draw_landmarks(frame, hand_landmarks, mp.solutions.hands.HAND_CONNECTIONS)
            #show image on the frame
            # Call victory_detection to draw rectangle based on landmarks
            frame = victory_detection(frame,hand_landmarks)
            frame = thumbup_detection(frame,hand_landmarks)
            frame = thumbdown_detection(frame,hand_landmarks)
            
    # Display the captured frame
    cv2.imshow('Real-Time Hand Detection App', frame)

    # Check for user input to exit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
